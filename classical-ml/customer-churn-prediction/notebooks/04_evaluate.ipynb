### Model Evaluation

```python
import pandas as pd
from tensorflow.keras.models import load_model
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Load Model and Test Data
print("Loading model and test data...")
model = load_model('../models/ann_model.keras')

X_test = pd.read_csv('../data/X_test_processed.csv')
y_test = pd.read_csv('../data/y_test.csv')['Exited']
```

```python
# Base Keras Evaluation
print("\n--- Base Keras Metrics ---")
metrics = model.evaluate(X_test, y_test)
for name, val in zip(model.metrics_names, metrics):
    print(f"{name.capitalize()}: {val:.4f}")
```

#### Threshold Optimization Analysis

```python
# Generate Predictions with Custom Threshold
threshold = 0.45
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > threshold).astype(int)

# Evaluation Metrics
print(f"\n--- Evaluation at Threshold: {threshold} ---")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
    
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
